```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """ 
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>
<body>
    <h1> asdfsdf </h1>
    <form action="/action">
        <label for="">
            Username
            <input type="text" id="User_Name" placeholder="Enter the text ">
        </label>
        <br>

        
        <br>
        Password
        <input type="password" placeholder="Enter you password">
        <br><br>
        <button>Submit karo</button>
        <button type="submit">Submit type</button>
        <button type="button">Normal button</button>
        <button type="reset">Reset type</button>

        <input type="button" value="Click me">
        <input type="submit">
        
        <input type="text" placeholder="id" name="Usename">
    </form>

    <form action="https://www.google.com/search">
        <input type="text" placeholder="Submit to google" >
        <button>search</button>
    </form>
</body>
</html>

"""

header_to_split_on = [
    ("h1", "Header 1"),
    # ("h2", "Header 2"),
    # ("h3", "Header 3")
]

html_splitter = HTMLHeaderTextSplitter (header_to_split_on)
html_header_splits = html_splitter.split_text(html_string)
html_header_splits
```
